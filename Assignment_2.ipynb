### Question-2

```python
# To calculate volume of sphere of unit radius
# centred at origin in the first quadrant

import math

n = 10000000
x = LCG.LCG(3, n)
y = LCG.LCG(5, n)
z = LCG.LCG(7, n)

count = 0
trial = 0

for i in range (0, n):
  p = math.sqrt(x[i]*x[i] + y[i]*y[i] + z[i]*z[i])
  if p <= 1:
    count = count + 1
    trial = trial + 1
  else:
    trial = trial + 1

volume = float(count/trial)
print("Volume of the sphere in the first quadrant = ", volume, "~ \u03C0/6")

               ######## OUTPUT ########
#Volume of the sphere in the first quadrant =  0.5239767 ~ Ï€/6
```

### Question-3

```python
import math
import matplotlib.pyplot as plt

# Defining a function to simulate random walk with given no. of steps

def random_walk(a: float, b: float, n: int):
  x0 = LCG.LCG(a, n)
  y0 = LCG.LCG(b, n)
  x1 = 0
  y1 = 0
  x = []
  y = []
  sum = 0
  
  for i in range(n):
    x1 += x0[i]
    x.append(x1)
    y1 += y0[i]
    y.append(y1)
    r = (x0[i]**2 + y0[i]**2)
    sum += r
  plt.plot(x, y)
  plt.title("Random walk with entered n steps")
  plt.show()
  rms = math.sqrt(sum/n)
  disp = math.sqrt(x0[n-1]**2 + y0[n-1]**2)
  print("rms distance for the walk = ", rms)
  print("net displacement for the walk = ", disp)


# Random walk with N = 300 steps
random_walk(2, 3, 300)

# Random walk with N = 600 steps
random_walk(5, 3, 600)

# Random walk with N = 900 steps
random_walk(3, 7, 900)
```
