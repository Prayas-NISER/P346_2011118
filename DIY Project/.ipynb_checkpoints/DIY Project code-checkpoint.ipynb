## Solving Laplace equation for Electrostatics numerically using Finite difference method

```python
# Creating a function for solving Laplace equation numerically using Finite Difference method

def Solve_Laplace_eq(t, b, l, r):
    import numpy as np
    import matplotlib.pyplot as plt

    # Setting maximum iteration
    max_iter = 1000
    # Setting dimensions and delta
    lenX = 50
    lenY = 50
    delta = 1
    # Setting boundary conditions
    Bound_top = t
    Bound_bottom = b
    Bound_left = l
    Bound_right = r
    # Initial guess of interior grid
    Tguess = 30

    # Setting colour interpolation and colour map
    color_intpol = 50
    colour_map = plt.cm.gist_rainbow

    # Creating grid for Finite difference method
    X, Y = np.meshgrid(np.arange(0,lenX), np.arange(0,lenY))

    # Setting array size and interior value equal to Tguess
    T = np.empty((lenX,lenY))
    T.fill(Tguess)

    T[(lenY-1):, :] = Bound_top
    T[:1, :] = Bound_bottom
    T[:, (lenX-1):] = Bound_right
    T[:, :1] = Bound_left

    # Solving using iterative method (Here, we assume that the solution converges within max_iter)
    for iter_ in range(max_iter):
        for i in range(1, lenX-1, delta):
            for j in range(1, lenY-1, delta):
                T[i, j] = 0.25 * (T[i+1][j] + T[i-1][j] + T[i][j+1] + T[i][j-1])


    # Plotting the contour
    plt.figure(figsize=(10,7))
    plt.title("2-D Contour of solution")
    plt.contourf(X, Y, T, color_intpol, cmap = colour_map)
    plt.colorbar()
    plt.show()
```

```python
# Showing the contour with boundary conditions:
# Top = 0, Bottom = 0, Left = 0, Right = 0
Solve_Laplace_eq(0, 0, 0, 0)
```
