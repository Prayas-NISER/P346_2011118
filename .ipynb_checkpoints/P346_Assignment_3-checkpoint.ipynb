```python
# To decompose a matrix using Cholesky Decomposition
import math
 
def Cholesky_Decomposition(matrix, n: int):
 
    lower = [[0 for x in range(n+1)]
                for y in range(n+1)];
 
    # Decomposing a matrix into Lower Triangular matrix
    for i in range(n):
        for j in range(i+1):
            sum1 = 0;
 
            # Summation for diagonals
            if (j == i):
                for k in range(j):
                    sum1 += pow(lower[j][k], 2);
                lower[j][j] = float(math.sqrt(matrix[j][j] - sum1));
            else: 
                # Evaluating L(i,j) using L(j,j)
                for k in range(j):
                    sum1 += (lower[i][k] * lower[j][k]);
                if(lower[j][j] > 0):
                    lower[i][j] = float((matrix[i][j] - sum1) / lower[j][j]);

    
    print("Lower Triangular matrix\t\tTranspose matrix");

    for i in range(n):
         
        for j in range(n):
            print(round(lower[i][j], 2), end = "\t");
        print("", end = "\t");
         
        for j in range(n):
            print(round(lower[j][i], 2), end = "\t");
        print("");
 
M1 = [[4, 12, -16],
      [12, 37, -43],
      [-16, -43, 98]]

Cholesky_Decomposition(M1, 3);
```

```python
# Solving a system of linear equations with Jacobi iterative method
def iterative_sol(A, B, X, e):
  sol1 = []
  sol2 = []
  sum = 0
  n = 0
  for r in range (2):
    for i in range(2):
      for j in range(2):
        if (i != j):
          sum += float(A[i][j] * X[r][0])
  for i in range(2):
    sol1[i][0] = float((B[i][0] - sum)/A[i][i])
  sum = 0
  for r in range (2):
    for i in range(2):
      for j in range(2):
        if (i != j):
          sum += float(A[i][j] * sol1[r][0])
  for i in range(2):
    sol2[i][0] = float((B[i][0] - sum)/A[i][i])
  sum = 0
  n += 1
  while abs(sol2[0][0] - sol1[0][0]) > e:
    for r in range (2):
      for i in range(2):
        for j in range(2):
          if (i != j):
            sum += float(A[i][j] * sol2[r][0])
    for i in range(2):
      sol2[i][0] = float((B[i][0] - sum)/A[i][i])
    sum = 0
    n += 1
  print(sol2)

A = [[4, -1, 1],
     [4, -8, 1],
     [-2, 1, 5]]

B = [[7],
     [-21],
     [15]]

X = [[0],
     [0],
     [0]]

iterative_sol(A, B, X, 0.001)


# Solving a system of linear equations with Gauss-Seidel iterative method
def gauss_seidel(A, B, n):
  X = [[0],
       [0],
       [0],
       [0]]
  sum_k1 = 0
  sum_k2 = 0
  for r in range(n):
    for i in range(4):
      for j in range(i+1,4):
        sum_k1 += A[i][j] * X[j][0]
      for j in range(i):
        sum_k2 += A[i][j] * X[j][0]
      X[i][0] = (B[i][0] - sum_k2 - sum_k1)/A[i][i]
      sum_k1 = 0
      sum_k2 = 0
  print(X)

A = [[4, 1, -1, 1],
     [1, 4, -1, -1],
     [-1, -1, 5, 1],
     [1, -1, 1, 3]]

B = [[-2],
     [-1],
     [0],
     [1]]

gauss_seidel(A, B, 100)
```

```python
#Function for reading the matrix from the given text file
def read_Mat(file_name):
    file=open(str(file_name),"r")
    M=[[]]
    temp=''
    i=0
    for j in file:
        for k in j:
            if k=='\n':
                M.append(int(temp))
                temp=''
                i+=1
                M+=[[]]
            elif k!=',' and k!='\n':
                temp+=k
            elif k==',' or k=='':
                M.append(int(temp))
                temp=''
    return(M)

#Function for printing the matrix in proper format
def print_Mat(mat,r,c):
    print("[",end="")
    for i in range(r):
        for j in range(c):
            temp = str(mat[i][j])
            if i!=0 and j == 0:
                print(" "+temp+"     ",end="")
            elif j==c-1:
                print(temp.rstrip(),end="")
            elif j!=c-1:
                print(temp+"     ",end="")
        if i!= r-1:
            print("\n")
        else:
            print("]")
    print()
```
