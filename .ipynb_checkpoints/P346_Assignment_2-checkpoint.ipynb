```python
# To calculate volume of sphere of unit radius
# centred at origin in the first quadrant

import math
n = 10000000
x = LCG.LCG(3, n)
y = LCG.LCG(5, n)
z = LCG.LCG(7, n)

count = 0
trial = 0

for i in range (0, n):
  p = math.sqrt(x[i]*x[i] + y[i]*y[i] + z[i]*z[i])
  if p <= 1:
    count = count + 1
    trial = trial + 1
  else:
    trial = trial + 1

volume = float(count/trial)
print("Volume of the sphere in the first quadrant = ", volume, "~ \u03C0/6")


               ######## OUTPUT ########
#Volume of the sphere in the first quadrant =  0.5239767 ~ π/6
```

```python
import matplotlib.pyplot as plt

# Defining a function to simulate random walk with given no. of steps
def random_walk(a: float, b: float, n: int):
  x0 = LCG.LCG(a, n)
  y0 = LCG.LCG(b, n)
  x1 = 0
  y1 = 0
  x = []
  y = []
  
  for i in range(n):
    x1 += x0[i]
    x.append(x1)
    y1 += y0[i]
    y.append(y1)
  plt.plot(x, y)
  plt.title("Random walk with entered n steps")
  plt.show()

random_walk(2, 3, 300)
random_walk(2, 3, 600)
random_walk(2, 3, 900)
```
